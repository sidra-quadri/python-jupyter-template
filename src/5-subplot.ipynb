### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
# Data for the plots
import numpy as np
x_line = np.linspace(0, 10, 100)
y_line = x_line**2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Line chart
axes[0, 0].plot(x_line, y_line, color='blue', label="y = x^2")
axes[0, 0].set_title("Line Chart")
axes[0, 0].set_xlabel("x")
axes[0, 0].set_ylabel("y")
axes[0, 0].legend()
axes[0, 0].grid(True)

# Bar chart
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")

# Scatter plot
axes[1, 0].scatter(x_scatter, y_scatter, color='orange', label="Data points")
for i in range(len(x_scatter)):
    axes[1, 0].text(x_scatter[i], y_scatter[i], f"({x_scatter[i]}, {y_scatter[i]})", fontsize=8)
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("x")
axes[1, 0].set_ylabel("y")
axes[1, 0].legend()
axes[1, 0].grid(True)

# Pie chart
axes[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=90, colors=['gold', 'skyblue', 'lightcoral', 'lightgreen'])
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()
plt.show()
```
